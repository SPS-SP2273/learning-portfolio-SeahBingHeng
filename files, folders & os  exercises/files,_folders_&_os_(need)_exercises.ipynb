```python
date_dict = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}
import os
import glob

def rename_my_file(old_file_name): 
   
    info = os.path.split(old_file_name)[-1] 
    month_date = info.split('_')[0].split('-')         
    month_value = date_dict[info.split('_')[0].split('-')[0]]
    new_format = f"{month_value}-{info.split('_')[0].split('-')[1]}-{info.split('_')[1]}"                                          
             
    return(new_format) 
    print(f'The old name {old_file_name} is now {new_format}')

txt_files = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
             'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
             'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
             'jan-01_tokyo.txt']

for txt in txt_files:                                                       
    new_name = rename_my_file(txt)
    print(f'{txt} ----> {new_name}')
    
```

# Task 7

```python
import os
import glob
import shutil

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']



for ct in cities:
    all_data = glob.glob(os.path.join('os-collab*',f'*{ct}.*'))
    #print(os.path.join('os-collab*',f'*{ct}.*'))

    for data in all_data:
        month_date_city_ext = os.path.split(data)[1]
        month_date_city = os.path.splitext(month_date_city_ext)[0]
        city = month_date_city.split('-')[-1]
    
        try:
            if city.capitalize() in cities:
                new_location = os.path.join('os-collaboration-exercise-data', city)        #eg os-collab-ex-data-Cairo. this is to put the file in to the Cairo folder
                if not os.path.exists(os.path.join(new_location, month_date_city_ext)):    #check whether the file is already in the folder. It joins the os-collab-ex-data-Cairo to the specific file eg 01-01-Cairo.txt
                    #shutil.move(data, new_location)               #move this file if it is not in the folder
                    print(f'Moved {data} to {new_location}')
                else:
                    print(f'File {month_date_city_ext} already exists in {new_location}')
        except PermissionError:
            print('PermissionError')
    
    # Check if all files were moved successfully
    for city in cities:
        moved_files = glob.glob(f'os-collaboration-exercise-data/{city}/*') #the f string has a {city} which allows me to find all the files in the specific city. EG /{city}/* lets me see all the files in the Cairo folder
        if len(moved_files) == len([data for data in all_data if city in data]):  #for data in all_data gets you the city in the file. if city in data means that the folder name aka Cairo is the same as city name in the file eg os-collab-01-01-Cairo.txt. 
            print(f"All {city} files moved successfully.")   #the first data creates a list of files that correspond to the folder name eg Cairo
        else:
            print(f"Some {city} files were not moved.")


```
