```python
dict = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}
import os
import glob

def rename_my_file(old_file_name): 
    info = old_file_name
    try:
        for months_key in dict:
            if info.split('_')[0].split('-')[0] == months_key:      #info.splitxxxx which is the month is the unknown, month_key is from the dict, takes all value 
                month_value = dict[info.split('_')[0].split('-')[0]]
                new_format = f"{month_value}-{info.split('_')[0].split('-')[1]}-{info.split('_')[1]}"  
                dire = f"os-collaboration-exercise-data"
                        
                new_new_form = os.path.join(dire, new_format)                     #get the os-collab + whatever numerical txt it is eg 04-01-sg.txt for the NEW code
                new_info = os.path.join(dire, old_file_name)                     #get os-collab + numerical text for OLD code 
               
                new_name = os.rename(new_info,new_new_form)                      #rename files names from old to name one, both including the file path os-collab xxxx            
                print(f'The old name {old_file_name} is now {new_format}')       #this shows the change w/o the file path                  
        
    except:
        print("File is not present")
        
old_file_name = input('what is the file name?')
rename_my_file(old_file_name)
```

# task 4


```python
dict = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}
import os
import glob

infos = glob.glob('os-collaboration*/*txt')

def rename_my_file(infos): 
    try:
        for information in infos:                      #information contains the individual list of txt files eg"apr-01_shanghai.txt, sep-01_singapore.txt"
            info = os.path.split(information)[1]              #we get the "month-day-country-ext" portion
            for months_key in dict:
                if info.split('_')[0].split('-')[0] == months_key:      #info.splitxxxx which is the month is the unknown, month_key is from the dict, takes all value 
                    month_value = dict[info.split('_')[0].split('-')[0]]
                    new_format = f"{month_value}-{info.split('_')[0].split('-')[1]}-{info.split('_')[1]}"  
                    dire = f"os-collaboration-exercise-data"
                                    
                    new_new_form = os.path.join(dire, new_format)                     #get the os-collab + whatever numerical txt it is eg 04-01-sg.txt for the NEW code
                    new_info = os.path.join(dire, info)                     #get os-collab + numerical text for OLD code 
                           
                    new_name = os.rename(new_info,new_new_form)                      #rename files names from old to name one, both including the file path os-collab xxxx            
                    print(f'The old name {info} is now {new_format}')       #this shows the change w/o the file path                  
            
    except:
            print("File is not present")
        
rename_my_file(infos)
```

# Task 7

```python
import os
import glob
import shutil
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
            'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

all_data = glob.glob('os-collab*/**')    #obtains all the files, including the folder that was created

for data in all_data:                       #data looks thru the list of files in all_data
    month_date_city_ext = os.path.split(data)[1]
    month_date_city = os.path.splitext(month_date_city_ext)[0]            #remove extension eg jpg
    city = month_date_city.split('-')[-1]                                 #split file name and choose last name which is city

    try:
        for all_cities in cities:                # 'all_cities' look at the data held in 'cities'
            if city.capitalize() == all_cities:
                new_location = os.path.join('os-collaboration-exercise-data', city)     #creating file path that connects the "os-exercise" file to city file
                shutil.copy(data,new_location)                   #copy the specific data (aka 04-01-singapore.txt) into the new file location eg os-collab-singapore
                print(f'Copied {data} to {new_location}')
    except:
        print('File is already moved into the right folder')
```
