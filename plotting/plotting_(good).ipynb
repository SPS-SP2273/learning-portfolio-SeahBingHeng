### pyplot Interface

```python
import numpy as np
from matplotlib import pyplot as plt
# Your code here
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
plt.plot(x, cos_x, label='cos x')                  #regular pyplot method
plt.plot(x, sin_x, label='sin x')
plt.legend()

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label='cos x')                           #object oreinted interface
ax.plot(x, sin_x, label='sin x')
ax.legend()
plt.show()
#pyplot interface for quick plots and the OO interface for more complex plots that demand control and finesse.
```

### OO Interface

# 3 What is OO ax, really?

```python

# Your code here
fig, ax = plt.subplots(nrows=2, ncols=1)          #to show the 2 diff plots in 2 rows and 1 column

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
plt.show()
```

# 4 A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=1,      #arrange plots in 2 rows 1 column
                       figsize=(5, 5),                #alter dimension of figure
                       sharex=True)                 #share x axis
ax[0].plot(x, cos_x, label=r'$\cos(x)$')             #plot cos graph, [0] and [1] is to specify the graph's location, eg cos is the first. possible as ax is an array
ax[0].fill_between(x, 0, cos_x, alpha=.25)      #fill is to shade the area under curve
ax[1].plot(x, sin_x, label=r'$\sin(x)$')            #plot sin graph
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for an in ax:                                # can individually do the labelling and axis etc, but use for loop to declutter code
    an.legend()
    an.grid(alpha=.25)                       #grindline opaqueness
    an.set_ylabel('$y$', rotation=0)           #y axis labelling

ax[1].set_xlabel('$x$')                  #x axis labelling dunnid go thru for loop since it is shared. Index 1 as we are labelling the x axis title below

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')            #title
fig.tight_layout()
plt.show()
```

# 5 Multiple rows and columns

### Using ax

```python

# Your code here
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey= 'col')       #subplot share x and y column. If set sharey as True, columns are shown in top left and bottom left graph
                                                        # as all graphs will have same y axis
# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')                #showing where the location of the titles can be placed
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)          #horizontal lines
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():                     #1D array eg [1,2,3,4] rather than [1,2,
                                                                               #3,4]
    a.grid(alpha=.25)

plt.tight_layout()
```

## 6.1 Histograms

### Data

```python
def get_plot_data():
    data = {}             
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)      #first row is header

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)          #look at the specific excel and transfer all data into data['NUS']

    return data
# Your code here
print(data['NUS'])    #list of all data related to NUS for visualisation
```

```python
plt.style.use('bmh')
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=60, label=['All', 'NUS'])    #create histogram from both data source and have 50 bins. Bins = 1 rectangle. less bins = larger range of x per rectangle to encompass more data
plt.xlabel('Mean of Basic Montly Earning (S$)')                         #eg when 5 bins, range can be like 0-500. When 100 bins, range can be like 0-10
plt.ylabel('Number of Students')
plt.legend()
```

## 6.2 Scatter plots

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items():    #label = either all or NUS
    print(len(numbers))  #show total number of data in all or NUS
    x = range(len(numbers))            #see how many sets of results are obtained per survey
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

## 6.3 Bar charts

```python
student_numbers = {'Life Sciences': 14,  #dictionary
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())     #get all key eg life science, physics
numbers = list(student_numbers.values())     #get all value eg 12 and 8
print(numbers)
plt.style.use('ggplot')
plt.bar(majors, numbers)           #use barh for horizontal plot
plt.xlabel('Majors')
plt.ylabel('Number of Students')
plt.show()
```

## 6.4 Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%100.3f%%',   # How to format the percentages. in the case of 1.3, '1' adjusts the position of the percentage while '3' is the number of DP
        startangle=100            
        )
plt.title('Percentage of each major')
```
