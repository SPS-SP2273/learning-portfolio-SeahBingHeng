```python

    #Remove the text I am..        DONE
    #Change the colours used for filling. DONE
    #Change the limits of the filled areas. DONE 
    #Add titles to each subplot. DONE
    #Share the x axis across columns. DONE 
    #Add/Remove labels to the x-axis. DONE
    #Make the tick labels of the two bottom plots the same.  DONE
    #Add grids to all subplots. DONE
    #Add a legend to all subplots in the upper right position. DONE
    #Use tight_layout() to improve the figure. DONE
    #Save the figure as a jpg file.

import numpy as np
from matplotlib import pyplot as plt
#--------- Generate cosine and sine values --------#
x = np.linspace(-np.pi, np.pi, num=100, endpoint=True)
cos_x = np.cos(x)
sin_x = np.sin(x)
fun1_x = np.exp(-x) * np.cos(5 * x)
fun2_x = np.exp(-x) * np.sin(2 * x)

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=2, ncols=2,
                         figsize=(12, 8),sharex= 'col',  sharey='row')

#------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='r', label='$\cos x$')
axes[0, 0].plot(x, cos_x**2, color='r',
                linestyle=':', label='$\cos^2 x$')
axes[0, 0].set_title('$\cos x$ & $\cos^2x$')
axes[0, 0].set_ylabel('Cosine Value')
axes[0, 0].fill_between(x, cos_x, -1, color='g', alpha=.125)
axes[0, 0].legend(loc='upper right', frameon=False)


#------- Subplot 2 -------#
axes[0, 1].set_title('$\sin x$ & $\sin^2x$')
axes[0, 1].plot(x, sin_x, color='g', label='$\sin x$')
axes[0, 1].fill_between(x, sin_x, -1, color='r', alpha=.125)
axes[0, 1].plot(x, sin_x**2, label='$\sin^2 x$')
axes[0, 1].set_ylabel('Cosine Value')
axes[0, 1].set_ylim(-1.25, 1.25)
axes[0, 1].legend(loc='upper right', frameon=False)

#------- Subplot 3 -------#
axes[1, 0].plot(x, fun1_x, color='b', label='$e^{-x}\cos 5x$')     #not sure if label shld be sin2x, according to the exercise picture
axes[1, 0].fill_between(x, fun1_x, -25, color='b', alpha=.125)
axes[1, 0].set_title('$e^{-x}\cos 5x$')
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_ylabel('Cosine Value')
axes[1, 0].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 0].set_xticklabels(['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$'])
axes[1, 0].legend(loc='upper right' )

#------- Subplot 4 -------#
axes[1, 1].plot(x, fun2_x, color='y', label='$e^{-x}\sin 2x$')        #not sure if label shld be sin2x, according to the exercise picture
axes[1, 1].set_title('$e^{-x}\sin 2x$')        
axes[1, 1].fill_between(x, fun2_x, -25, color='y', alpha=.125)
axes[1, 1].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 1].set_xticklabels(['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$'])
axes[1, 1].set_xlabel('Angle (radians)')
axes[1, 1].legend()

# 'flatten', 'opens' the 2D array into a simple 1D array
for a in axes.flatten():
        a.grid(alpha=.5)

# plt.tight_layout()
fig.tight_layout()

plt.savefig('exercise plotting.jpg', dpi=150)     #save pic
```
