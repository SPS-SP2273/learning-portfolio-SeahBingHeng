```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color= 'red', linestyle= 'dashed',         #colour and style of the normal graph along with how the line shld be drawn
             marker= 'o' , markerfacecolor= 'yellow',          #marker is the style of each error plot. Can be circular, triangular etc. markerfacecolour is the colour of the marker, so it is yellow here
             ecolor= 'black',capsize=5,                 #error bar line colour. capsize is the length of the error bar
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc ='lower right')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
    return np.exp(-x/10)*np.sin(x)

x = np.linspace(0, 50, 100)             #set the x axis range from 0 to 50, with 100 items in between for a smooth curve
y = f(x)


plt.plot(x,y,linestyle = 'solid',color = 'blue')
plt.title(r'y = $e^{-x/10}sin(x)$', fontsize = 20) 
plt.xlabel('x', color = 'darkorange', fontsize = 15)
plt.ylabel('y', color = 'darkorange', fontsize = 15) 
plt.ylim(-1,1)                     #cannot use ylim = [-1,1] as it doesnt include the max and min value stated
plt.grid(alpha = 0.25)
plt.show()
```
